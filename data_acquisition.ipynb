```python
import requests

url = "https://api.themoviedb.org/3/genre/movie/list?api_key=9cc45ac27212800abff00a0e31cce03c&language=en-US"


response = requests.request("GET", url)

print(response.text)
```

```python
genre_ids = []
for genre in response.json()['genres']:
    genre_ids.append(genre['id'])
```

```python
genre_ids
```

```python
import tmdbsimple as tmdb
tmdb.API_KEY = '9cc45ac27212800abff00a0e31cce03c'
tmdb.REQUESTS_TIMEOUT = 5  # seconds, for both connect and read
```

```python
from tqdm import tqdm
import pandas as pd
discover = tmdb.Discover()
for genre_id in genre_ids:
    results = []
    print(genre_id)
    for i in tqdm(range(300)):
        response = discover.movie(with_genres=genre_id,page=i+1)
        for temp in response['results']:
            if 'release_date' in temp.keys():
                results.append([temp['id'],temp['genre_ids'],temp['title'],temp['original_language'],temp['overview'],temp['popularity'],temp['poster_path'],temp['vote_average'],temp['release_date']])
            else:
                results.append([temp['id'],temp['genre_ids'],temp['title'],temp['original_language'],temp['overview'],temp['popularity'],temp['poster_path'],temp['vote_average'],None])
    df = pd.DataFrame(results,columns=['id','genre_ids','title','language','overview','popularity','poster_path','vote_average','release_date'])
    df.to_csv(f'{genre_id}.csv',index=False)
```

```python
import os,base64
import requests as req
from PIL import Image
from io import BytesIO
from tqdm import tqdm
import pandas as pd

flag = 0
for genre in genre_ids[1:]:
    if genre == 36:
        flag = 1
    if flag == 0:
        continue
    os.mkdir(f'{genre}')
    df = pd.read_csv(f'{genre}.csv')
    print(genre)
    for _,s in tqdm(df.iterrows()):
        try:
            path = s['poster_path']
            response = requests.get(f'http://image.tmdb.org/t/p/w500/{path}')
            if response.status_code != 404:
                image = Image.open(BytesIO(response.content))
                ls_f=base64.b64encode(BytesIO(response.content).read()).decode('utf-8')
                imgdata=base64.b64decode(ls_f)
                movie_id = s['id']
                file=open(f'{genre}/{movie_id}.jpg','wb')
                file.write(imgdata)
                file.close()
        except Exception:
            continue
```
